```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the dataset
dataset = pd.read_csv('linear.csv')
dataset.head()
```

```python
plt.scatter(dataset['cgpa'], dataset['package'])
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.title('CGPA vs Package')
plt.show()
```

```python
# separate x and y
x = dataset.iloc[:,0:1] # input
y = dataset.iloc[:,-1] # output
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
# Create a linear regression model
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_test
```

```python
y_test
```

```python
# Predict the output for the test set
model.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
# Show Regression line
plt.scatter(dataset['cgpa'], dataset['package'])
plt.plot(x_train, model.predict(x_train), color='red')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.title('CGPA vs Package')
plt.show()
```

```python
# Now see that how that line is working on test data
# Y = mX + c

m = model.coef_
c = model.intercept_
print(f"Coefficient (m): {m}")
print(f"Intercept (c): {c}")
```

```python
# Now we can predict the package for a new CGPA value

new_cgpa = 9.0
predicted_package = m * new_cgpa + c
print(f"Predicted package for CGPA {new_cgpa}: {predicted_package}")
```
